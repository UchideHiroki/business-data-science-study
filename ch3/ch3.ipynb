```python
# FDRに基づいて説明変数を選択する
q = 0.1
coef_p_bh = pd.DataFrame({
    "coef": res.params.index,
    "p": res.pvalues
})
coef_p_bh = coef_p_bh.query("coef != 'Intercept'")
coef_p_bh.sort_values(["p"], inplace=True)
coef_p_bh.reset_index(drop=True, inplace=True)
coef_p_bh["order"] = list(range(1, coef_p_bh.shape[0]+1))
coef_p_bh["bh"] = [q * (k/coef_p_bh.shape[0]) for k in range(1, coef_p_bh.shape[0]+1)]
coef_p_bh["is_accept"] = "reject"
max_accept_index = coef_p_bh[coef_p_bh.p <= coef_p_bh.bh].index.max()
coef_p_bh.loc[0:max_accept_index, "is_accept"] = "accept"
```

```python
trace1, trace2 = px.scatter(coef_p_bh, x="order", y="p", color="is_accept", color_discrete_map={"accept": "red", "reject": "gray"}, hover_name="coef").data
```

```python
trace3 = px.line(coef_p_bh, x="order", y="bh").data[0]
```

```python
fig = go.Figure()
fig.add_traces([trace1, trace2, trace3])
fig.update_layout(height=400, width=600)
fig.update_xaxes(type="log")
fig.update_yaxes(type="log")
fig.show()
```

25個の説明変数を選択した。  
選択した説明変数のみで簡素なモデルを作る。

```python
formula = "FAIL ~ " + "+".join(coef_p_bh.query("is_accept == 'accept'").coef)
cut = glm(formula=formula, data=SC, family=sm.families.Binomial()).fit()
```

```python
print(cut.summary())
```

```python
1 - cut.deviance/cut.null_deviance
```

決定係数は下がった。(自由度調整していないし当たり前)

```python
# せっかくなので、自由度調整済み決定係数を計算してみる
full_r2_adjusted = (res.deviance/res.df_resid)/(res.null_deviance/(SC.shape[0]-1))
cut_r2_adjusted = (cut.deviance/cut.df_resid)/(cut.null_deviance/(SC.shape[0]-1))
print(full_r2_adjusted, cut_r2_adjusted)
```

後者の方が断然説明力がある

```python
def deviance(y, pred, family="binomial"):
    assert family in (["binomial", "gaussian"]), "family should be binomial or gaussian"

    if family == "binomial":
        return -2 * np.sum(y * np.log(pred) + (1 - y) * np.log(1 - pred))
    else:
        return np.sum((y - pred)**2)

def R2(y, pred, family="binomial"):
    dev = deviance(y, pred, family)
    null_dev = deviance(y, np.mean(y), family)
    return 1 - dev/null_dev
```

```python
# sanity check
R2(SC.FAIL, cut.predict(SC)), 1 - cut.deviance / cut.null_deviance
```

```python
formula_full = "FAIL ~ " + "+".join([f"SIG{i}" for i in range(1, 201)])
formula_cut = "FAIL ~ " + "+".join(coef_p_bh.query("is_accept == 'accept'").coef)
```

```python
kf = KFold(n_splits=10, random_state=1234, shuffle=True)
r2full_list = []
r2cut_list = []

for train_index, test_index in kf.split(SC):
    rfull = glm(formula=formula_full, data=SC.loc[train_index], family=sm.families.Binomial()).fit()
    rcut = glm(formula=formula_cut, data=SC.loc[train_index], family=sm.families.Binomial()).fit()
    
    predfull = rfull.predict(SC.loc[test_index])
    predcut = rcut.predict(SC.loc[test_index])

    r2full_list.append(R2(SC.loc[test_index].FAIL, predfull))
    r2cut_list.append(R2(SC.loc[test_index].FAIL, predcut))
```

```python
oos_r2_df = pd.DataFrame({
    "full": r2full_list,
    "cut": r2cut_list
}).melt(value_vars=["full", "cut"], var_name=["type"], value_name="R2")
```

```python
px.box(oos_r2_df, x="type", y="R2", height=400, width=500)
```

FDRに基づいた変数選択の問題点  
多重共線性があると、相関のある全ての変数のp値が大きくなる、どれを含めるべきかわからない  
p値は過剰適合したモデルから得られるので信頼性に欠ける  

変数減少法よりも、変数増加法(AICなどを基準に変数を追加していく)の方がいい  
正則化の考え方を取り入れると劇的に改善できる(らしい)

```python
from scipy.sparse import csr_matrix
from sklearn.linear_model import LassoCV, LassoLarsIC
from sklearn.preprocessing import StandardScaler
```

```python
X = csr_matrix((browser_domains.visits, (browser_domains.id - 1, browser_domains.site - 1)))
```

```python
X.shape
```

```python
scaler = StandardScaler(with_mean=False)
```

```python
X_std = scaler.fit_transform(X)
```

```python
y = np.log(browser_totalspend.spend)
```

```python
reg = LassoCV(cv=5, n_alphas=20, copy_X=False, verbose=True, normalize=False, n_jobs=-1, random_state=1234).fit(X=X_std, y=y)
```

対数λと、mseをプロットしてみる

```python
lambda_mse = pd.DataFrame({
    "log_lambda": np.log(reg.alphas_),
    "mse": np.mean(reg.mse_path_, axis=1)
})
```

```python
px.scatter(lambda_mse, x="log_lambda", y="mse", range_x=[-6.5, -1.5], height=400, width=500)
```
